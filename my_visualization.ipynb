```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
my_x = np.arange(1,10,1)
```

```python
my_x
```

```python
my_y = np.arange(1,10,1)
```

```python
%matplotlib inline
```

```python
plt.plot(my_x,my_y,'r-')
plt.xlabel('My X')
plt.ylabel('My Y')
plt.title('My Plot')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
#axes2 = fig.add_axes([0.4,0.4,0.4,0.3])
axes1.plot(my_x,my_y)
axes1.set_xlabel('My X')
axes1.set_ylabel('My X')
axes1.set_title('My X')
```

```python
my_arr1 = np.linspace(0,10)
my_arr2 = my_arr1 ** 2
plt.plot(my_arr1, my_arr2)
```

```python
ax = plt.figure()
fig = ax.add_axes([0,0,1,1])
fig.plot(my_arr1,my_arr2)
fig.plot(my_arr2,my_arr1)

fig2 = ax.add_axes([1,0,1,1])
fig2.scatter(my_arr1,my_arr2)
```

```python
ax,axes = plt.subplots(1,3,figsize=(15,4))
#fig = ax.add_axes([0,0,1,1])
axes[0].plot(my_arr1,my_arr2)
axes[0].plot(my_arr2,my_arr1)

#fig2 = ax.add_axes([1,0,1,1])
axes[1].scatter(my_arr1,my_arr2)

axes[2].hist(my_arr1,my_arr2,ls='--',color='orange')
axes[2].set_xlim([0,10])

ax
```
